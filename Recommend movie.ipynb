## Converting string of list to list

```python
import ast

def converter(obj):
    # Check if obj is a string, if so, use ast.literal_eval to convert
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    
    # Assuming obj is a list of dictionaries, extract 'name'
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
def cast_converter(obj):
    # Check if obj is a string, if so, use ast.literal_eval to convert
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    
    # Assuming obj is a list of dictionaries, extract 'name'
    L = []
    val = 0
    for i in obj:
        val = val + 1
        L.append(i['name'])
        if val>=3:
            break
    return L
```

```python
movies.iloc[:, 1] = movies['cast'].apply(cast_converter)
```

```python
def crew_converter(obj):
    # Check if obj is a string, if so, use ast.literal_eval to convert
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    
    # Assuming obj is a list of dictionaries, extract 'name'
    L = []
    for i in obj:
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

## Text Vectorization for similarity calculation

For recommendation we suggest movies based on their similarity, and similarity is calculated based on generated tags.
For this we need to vectorize the tags first and after converting them into a vector we can perform calculation to find similarity between any 2 movies.
Based on this we will suggest the movie to the user.

## Different techniques for text vectorization
    
1. Bag of words (I am using this one)
2. TF IDF
3. word2vec

## Working of bag of words

The Bag of Words (BoW) model is a simple way to convert text into numerical data for machine learning. Here's how it works in simple steps:

  ### 1. Combing words
  All words in the entire dataset is merge into a bag of words
  
  ### 2. calculating frequency
  Calculate frequency for each word in the bag of words for every movie
  
  ### 3. Vectorization
The frequency for each word is considered a vector for that movie  

### 4. Ignoring stop words
All stops words like but, in, a, the etc are not considered for this frequency calculation because they have no meaning in calculating similarity between movies.
  

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV = CountVectorizer(max_features = 5000, stop_words = 'english')
```

## Calculating similarity through cosine similarity

I have calculated cosine similarity between every vectors to mesaure its similarity with one another

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
import pickle

# Save the similarity matrix to a file
with open("similarity.pkl", "wb") as file:
    pickle.dump(similarity, file)
print("Similarity matrix saved successfully!")
```

```python
# Load the similarity matrix from the file
with open("similarity.pkl", "rb") as file:
    similarity = pickle.load(file)
print("Similarity matrix loaded successfully!")
```

```python
def recommend(movie):
    # Converting the input movie title to lowercase
    movie = movie.lower()

    # Check if the movie exists in the DataFrame
    if movie not in final_df['title'].values:
        print(f"'{movie.title()}' not found in the database.")
        return

    # Find the index of the movie in the final DataFrame
    index = final_df[final_df['title'] == movie].index[0]
    
    # Get the similarity scores of this movie with all other movies
    similarity_with_other_movies = similarity[index]
    
    # Sort the similarity scores in descending order and get the top 5 most similar movies (ignoring the movie itself)
    similar_movies = sorted(list(enumerate(similarity_with_other_movies)), key=lambda x: x[1], reverse=True)[1:6]
    
    # Print the titles of the recommended movies in capatalized format
    for i in similar_movies:
        print(final_df.iloc[i[0]]['title'].title())
```

## Explanation of Each Line:

`def recommend(movie):`

Defines a function `recommend` that takes one argument, `movie`, which is the title of the movie you want to find recommendations for.

---

`index = final_df[final_df['title'] == movie].index[0]`

- This line searches for the row in `final_df` where the 'title' column matches the given movie title.
- It retrieves the index of that row using `.index[0]`, which is the index of the first movie found.
- Example: If `movie = 'Avatar'`, and 'Avatar' exists at index 1 in `final_df`, this line will return `1`.

---

`similarity_with_other_movies = similarity[index]`

- `similarity` is assumed to be a precomputed similarity matrix (e.g., cosine similarity).
- This line retrieves the similarity scores between the given movie and all other movies using the index we found earlier.
- The result is a list or array where each element represents how similar another movie is to the one at the index position.

---

`similar_movies = sorted(list(enumerate(similarity_with_other_movies)), key=lambda x: x[1], reverse=True)[1:6]`

- `enumerate(similarity_with_other_movies)`: This creates a list of tuples, where each tuple is of the form `(index, similarity_score)` for every movie in the similarity matrix.
- `sorted(..., key=lambda x: x[1], reverse=True)`: This sorts the list of tuples by similarity scores (`x[1]`), in descending order (highest similarity first).
- `[1:6]`: This slices the sorted list to get the 5 most similar movies, skipping the first one (`[1]`), because the highest similarity is the movie itself (similarity with itself is 1).
- **Result**: You now have a list of the 5 most similar movies (their indices and similarity scores).

---

`for i in similar_movies:`

- Iterates over the list of similar movies. Each `i` is a tuple containing the index of a movie and its similarity score (e.g., `(3, 0.85)`).

---

`print(final_df.iloc[i[0]]['title'])`

- For each similar movie, it retrieves the movie title using `.iloc[i[0]]`, where `i[0]` is the index of the similar movie.
- `final_df.iloc[i[0]]['title']`: This looks up the movie title at the index `i[0]` and prints it.
